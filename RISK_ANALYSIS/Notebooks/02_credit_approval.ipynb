```python
import pandas as pd
import numpy as np
# 加载数据
df=pd.read_csv('order_table_50k_v2.csv')
```

```python
##$2000
df['cs1']=np.where(((df.principal<2000))&(df.control_group==0),1,0)
df_pivot=df[df['control_group']==0].groupby("cs1").agg(order_cnt=("order_id", "size"),
                         approval_cnt=("order_status", "sum"),
                         fpd30=("fpd30", "sum"),
                         mob3_m2=("mob3_m2", "sum"),
                         principal=("principal", "sum")).reset_index()
df_pivot['fpd30_rate']=df_pivot['fpd30']/df_pivot['approval_cnt']
df_pivot['mob3_m2_rate']=df_pivot['mob3_m2']/df_pivot['approval_cnt']
df_pivot
```

```python
##2009
df['cs1']=np.where(((df.age<21)|(df.fico_score<600))&(df.control_group==0),1,0)
df_pivot=df[df['control_group']==0].groupby("cs1").agg(order_cnt=("order_id", "size"),
                         approval_cnt=("order_status", "sum"),
                         fpd30=("fpd30", "sum"),
                         mob3_m2=("mob3_m2", "sum"),
                         principal=("principal", "sum")).reset_index()
df_pivot['fpd30_rate']=df_pivot['fpd30']/df_pivot['approval_cnt']
df_pivot['mob3_m2_rate']=df_pivot['mob3_m2']/df_pivot['approval_cnt']
df_pivot.to_csv('2009_result.csv')
```

```python
##2009 添加金额逾期率
##2009
df['cs1']=np.where(((df.age<21)|(df.fico_score<600))&(df.control_group==0),1,0)
df['fpd30_prin']=df['principal']*df['fpd30']
df['mob3_m2_prin']=df['principal']*df['mob3_m2']
df_pivot=df[df['control_group']==0].groupby("cs1").agg(order_cnt=("order_id", "size"),
                         approval_cnt=("order_status", "sum"),
                         fpd30=("fpd30", "sum"),
                         mob3_m2=("mob3_m2", "sum"),
                         principal=("principal", "sum"),
                         fpd30_prin=("fpd30_prin", "sum"),
                         mob3_m2_prin=("mob3_m2_prin", "sum")).reset_index()
df_pivot['fpd30_rate']=df_pivot['fpd30']/df_pivot['approval_cnt']
df_pivot['mob3_m2_rate']=df_pivot['mob3_m2']/df_pivot['approval_cnt']
df_pivot['fpd30_prin_rate']=df_pivot['fpd30_prin']/df_pivot['principal']
df_pivot['mob3_m2_prin_rate']=df_pivot['mob3_m2_prin']/df_pivot['principal']
df_pivot
```
